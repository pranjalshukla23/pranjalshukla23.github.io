```python

from flask import Flask, request
from final_year_project_Ratings_DecisionTree_interface import predictRating
from flask import render_template


app = Flask(__name__)
app.config["DEBUG"] = True

@app.route("/", methods=["GET", "POST"])
def adder_page():
    errors = ""
    if request.method == "POST":
        number1=None
        number2=None
        try:
            number1 = int(request.form["matches"])
        except:
            errors += "<p>{!r} is not a number.</p>\n".format(request.form["matches"])
        try:
            number2 = int(request.form["id"])
        except:
            errors += "<p>{!r} is not a number.</p>\n".format(request.form["id"])
       
        
        if number1 is not None and number2 is not None:
            result = predictRating(number1,number2)
            return '''
                <html>
                    <body>
                        <p>The rating  is {result}</p>
                        <p><a href="/">Click here to calculate again</a><br>
                       </body>
                </html>
            '''.format(result=result)

    return '''
        <html>
            <body>
                {errors}
                <p>Enter your information:</p>
                <form method="post" action=".">
                    <label for="matches">Enter the matches</label>
                    <input type="text" name="matches"/><br>
                    <label for="id">Enter the team id:</label>
                    <input name="id" type="text"/><br>
                    <p><input type="submit" value="Predict the rating" /></p>
                </form>
            </body>
        </html>
    '''.format(errors=errors)
if __name__== "__main__":
    from werkzeug.serving import run_simple
    run_simple('localhost', 10200, app)

```
