```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
values=[52,79,43,77]
items=["Linear R2","Random R2  ","linear accuracy","random accuracy"]
plt.yticks(np.arange(0,1000,5))
plt.xlabel("parameters")
plt.ylabel("values")
plt.bar(items,values,color=["blue","red"])
plt.title("ODI data analysis")
plt.show()
```
